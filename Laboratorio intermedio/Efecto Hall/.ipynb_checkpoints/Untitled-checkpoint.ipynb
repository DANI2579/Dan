```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize as scp
```

```python
w = 1e-3
```

```python
def regresion(x, y, x_axis, y_axis, title, x_err, y_err, leyenda_ajuste,leyenda_datos, colors):
    pendiente =[]
    corte = []
    sigma_p = []
    sigma_c =[]
    fig, axs = plt.subplots(2,1, sharex=False,figsize=(6,7), gridspec_kw={'height_ratios': [2, 1]})
    plt.subplots_adjust(hspace=0)
    for i in range(0, len(x)):
        coefficients,cov=np.polyfit(x[i], y[i], 1, cov=True)
        poly=np.poly1d(coefficients)
        y_trend = poly(x)
    
        residuals= y[i]- poly(x[i])
        normalised_residuals=residuals/np.std(residuals)
        m, b = coefficients[0], coefficients[1]
        sigma_m, sigma_b = np.sqrt(cov[0,0]), np.sqrt(cov[1,1])
        pendiente.append(m)
        corte.append(b)
        sigma_p.append(sigma_m)
        sigma_c.append(sigma_b)
    
        #Grafica
        
        axs[0].plot(x[i], y_trend[i], 
                    label=leyenda_ajuste[i], ls="-", c=colors[i][0])
        axs[0].errorbar(x[i], y[i], yerr =y_err[i], 
                        xerr = x_err[i], fmt="o", label=leyenda_datos[i], 
                        ecolor=colors[i][1], capsize=3,c=colors[i][1], markersize=5)
        axs[0].set_title(title)
        axs[0].set_ylabel(y_axis)
        axs[0].legend(loc=0)
        axs[0].grid("major")
        
        
        axs[1].axhline(y=0,color='k',linestyle='--')
        axs[1].plot(x[i], normalised_residuals, linestyle='None', marker='o', 
                    markerfacecolor='none', markeredgecolor= colors[i][1])
        axs[1].grid("major")
        axs[1].set_xlabel(x_axis)
        axs[1].set_ylabel("Residuos normalizados")
    
    plt.savefig(f"{title}.jpeg")

    table = {"Pendiente": pendiente, "Punto de corte":corte, "$\sigma_m$":sigma_p, "$\sigma_b$":sigma_c}

    table = pd.DataFrame(table)

    return table
    
        
```

### Actividad 3

```python
def modelo (x, a, b, l):
    return a* (x) ** b + l
```

```python
# Tipo p
datos_act_3 = pd.read_excel("Datos efecto hall.xlsx", 
                            sheet_name="Actividad 3 p", skiprows=1)
```

```python
# Actividad 3.1
current_p_B0 = (datos_act_3["I_p(mA)"][0].
            astype(float))
long_voltage_p_B0 = (datos_act_3["V_l(V)"][0].
            astype(float))
R0 = long_voltage_p_B0/current_p_B0
sigma_R0 = 0.05/ current_p_B0 * np.sqrt(1 + R0**2)
```

```python
# Actividad 3.2
current_p_B = (datos_act_3["Unnamed: 2"][1:].
            astype(float).to_numpy())
long_voltage_p_B = (datos_act_3["Unnamed: 3"][1:].
            astype(float).to_numpy())
R = long_voltage_p_B/current_p_B0
sigma_R =0.05/ current_p_B * np.sqrt(1 + R**2)
```

```python
yp = (R-R0)/R0
sigma_yp = np.sqrt((1/R0)**2 * sigma_R**2 + (R/R0)**2* sigma_R0)

current = current_p_B

m_array = np.full(current.shape, m, dtype=float)
b_array = np.full(current.shape, b, dtype=float)
sigma_m_array = np.full(current.shape, sigma_m, dtype=float)
sigma_b_array = np.full(current.shape, sigma_b, dtype=float)
delta_I = np.full(current.shape, 0.05, dtype=float)

B_p = m_array * current + b_array

sigma_B_p = np.sqrt((current * sigma_m_array)**2 + sigma_b_array**2 + (m_array * delta_I)**2)
```

```python
# Tipo p
datos_act_3 = pd.read_excel("Datos efecto hall.xlsx", 
                            sheet_name="Actividad 3 n")
```

```python
current_n_B0 = (datos_act_3["I_p(mA)"][0].
            astype(float))
long_voltage_n_B0 = (datos_act_3["V_l(V)"][0].
            astype(float))
R0 = long_voltage_n_B0/current_n_B0
sigma_R0 = 0.05/ current_n_B0 * np.sqrt(1 + R0**2)
```

```python
# Actividad 3.2
current_n_B = (datos_act_3["I(A)"].
            astype(float).to_numpy())
long_voltage_n_B = (datos_act_3["V_l(V).1"].
            astype(float).to_numpy())
R = long_voltage_n_B/current_n_B0
sigma_R = 0.05/ current_n_B0 * np.sqrt(1 + R**2)
```

```python
yn = (R-R0)/R0
sigma_yn = np.sqrt((1/R0)**2 * sigma_R**2 + (R/R0)**2* sigma_R0)


current = current_n_B

m_array = np.full(current.shape, m, dtype=float)
b_array = np.full(current.shape, b, dtype=float)
sigma_m_array = np.full(current.shape, sigma_m, dtype=float)
sigma_b_array = np.full(current.shape, sigma_b, dtype=float)
delta_I = np.full(current.shape, 0.05, dtype=float)

B_n = m_array * current + b_array

sigma_B_n = np.sqrt((current * sigma_m_array)**2 + sigma_b_array**2 + (m_array * delta_I)**2)
```

```python
popt_p, pcov_p = scp.curve_fit(modelo, B_p, yp)
popt_n, pcov_n = scp.curve_fit(modelo, B_n, yn)
```

```python
y_model_p = modelo(B_p, *popt_p)
y_model_n = modelo(B_n, *popt_n)
```

```python
residuals_p= yp- y_model_p
normalised_residuals_p=residuals_p/np.std(residuals_p)
residuals_n= yn- y_model_n
normalised_residuals_n=residuals_n/np.std(residuals_n)

a_p, b_p, l_p = popt_p[0], popt_p[1], popt_p[2]
a_n, b_n, l_n = popt_n[0], popt_n[1], popt_n[2]
sigma_a_n, sigma_b_n, sigma_l_n = np.sqrt(pcov_n[0,0]), np.sqrt(pcov_n[1,1]), np.sqrt(pcov_n[2,2])
sigma_a_p, sigma_b_p, sigma_l_p = np.sqrt(pcov_p[0,0]), np.sqrt(pcov_p[1,1]), np.sqrt(pcov_p[2,2])
```

```python
#Grafica
fig, axs = plt.subplots(2,1, sharex=False,figsize=(6,7), gridspec_kw={'height_ratios': [2, 1]})
plt.subplots_adjust(hspace=0)
axs[0].plot(B_p, y_model_p, 
             ls="-", c="#4E79A7")
axs[0].errorbar(B_p, yp, 
                xerr = sigma_B_p, fmt="o", label="Tipo p", 
                ecolor="#4E79A7", capsize=3,c="#4E79A7", markersize=5)
axs[0].plot(B_n, y_model_n, 
             ls="-", c="#F28E2B")
axs[0].errorbar(B_n, yn, 
                xerr = sigma_B_n, fmt="o", label="Tipo n", 
                ecolor="#F28E2B", capsize=3,c="#F28E2B", markersize=5)

axs[0].set_ylabel("$R-R_0/R_0$")
axs[0].legend(loc=0)
axs[0].grid("major")


axs[1].axhline(y=0,color='k',linestyle='--')
axs[1].plot(B_p, normalised_residuals_p, linestyle='None', marker='o', 
            markerfacecolor='none', markeredgecolor= "#4E79A7")
axs[1].plot(B_n, normalised_residuals_n, linestyle='None', marker='o', 
            markerfacecolor='none', markeredgecolor= "#F28E2B")
axs[1].grid("major")
axs[1].set_xlabel("Campo magn√©tico (mT)")
axs[1].set_ylabel("Residuos normalizados")
plt.savefig("Magnetoresistencia.jpeg")
```
